```python
import pandas as pd
df = pd.read_csv("Food_Time new.csv")

df.info()
```

```python
df = df.dropna(subset=["weather_description"])
```

```python
df.isnull().sum()
```

```python
df_median = df["Distance (km)"].median()
```

```python
df["Distance (km)"] = df["Distance (km)"].fillna(df_median)
```

```python
df.isnull().sum()
```

```python
df["Restaurant_longitude"] = (
    df["Restaurant_longitude"]
    .str.replace(".", "", regex=False)   # rimuove i punti
    .astype(int)                         # converte in intero
    / 1_000_000                          # converte in gradi decimali
)

df["Restaurant_latitude"] = (
    df["Restaurant_latitude"]
    .str.replace(".", "", regex=False)   # rimuove i punti
    .astype(int)                         # converte in intero
    / 1_000_000                          # converte in gradi decimali
)

df["Delivery_location_latitude"] = (
    df["Delivery_location_latitude"]
    .str.replace(".", "", regex=False)   # rimuove i punti
    .astype(int)                         # converte in intero
    / 1_000_000                          # converte in gradi decimali
)

df["Delivery_location_longitude"] = (
    df["Delivery_location_longitude"]
    .str.replace(".", "", regex=False)   # rimuove i punti
    .astype(int)                         # converte in intero
    / 1_000_000                          # converte in gradi decimali
)


```
