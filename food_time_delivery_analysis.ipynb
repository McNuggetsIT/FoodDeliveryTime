```python
def convert_time(x):
    try:

        x = str(x)

        if x.count('.') > 1:
            x = x.replace('.', '')

        num = float(x)

        if len(x) > 5:
            return round(num / 60000000, 2)
        else:
            return num
    except ValueError:
        return None

handler.df['target'] = handler.df['target'].apply(convert_time)
```

```python
def fix_coord(s: str) -> float:
    # Remove all dots
    s_clean = s.replace('.', '')
    # Convert to float and put decimal in correct place
    return float(s_clean) / 1_000_000

handler.df['restaurant_latitude'] = handler.df['restaurant_latitude'].apply(fix_coord) 
handler.df['restaurant_longitude'] = handler.df['restaurant_longitude'].apply(fix_coord) 
handler.df['delivery_location_latitude'] = handler.df['delivery_location_latitude'].apply(fix_coord) 
handler.df['delivery_location_longitude'] = handler.df['delivery_location_longitude'].apply(fix_coord) 
```
